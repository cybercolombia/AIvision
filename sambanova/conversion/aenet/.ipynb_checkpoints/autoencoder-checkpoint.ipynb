## Autoencoder
An autoencoder is designed to compress the information of an input $x_i$ by passing it through an encoder set of layers $g(x)$, until it gets to the lower dimensional bottleneck or latent space (the code $z_i=g(x_i)$) with reduced information. The information in the latent space $z_i$ is then passed through the decoder $h(z)$, which should restore most of the original input $\bar{x}_i=h(z_i)$. 

![](autoencoder_arch.png)

```python
import torch
import torch.nn as nn
from torchvision import datasets
from torchvision import transforms

import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using {device} device')
```

```python
#Set seed for reproducibility
def seeding(seed):
    #random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
```

```python
seeding(123)
```

### Load dataset

```python
image_path = './'
batch_size_train = 32
batch_size_test = 1000
```

```python
tensor_transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root=image_path,
                            train=True,
                            download=True,
                            transform=tensor_transform)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                        batch_size=batch_size_train,
                                        shuffle=True)
```

```python
test_dataset = datasets.MNIST(root=image_path,
                            train=False,
                            download=True,
                            transform=tensor_transform)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                        batch_size=batch_size_test,
                                        shuffle=True)
```

### Create the autoencoder
The encoder will take the dimensions from 28*28 to 9 through a series of layers:

784 ==> 128 ==> 64 ==> 36 ==> 18 ==> 9

And then the decoder will take it back:

9 ==> 18 ==> 36 ==> 64 ==> 128 ==> 784

```python
class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()

        # Encoder with linear layers follower by ReLU activations
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 36),
            nn.ReLU(),
            nn.Linear(36, 18),
            nn.ReLU(),
            nn.Linear(18, 9)
        )

        # Decoder with linear layers followed by ReLU activations.
        # At the end a sigmoid function outputs a value between 0 and 1
        self.decoder = nn.Sequential(
            nn.Linear(9, 18),
            nn.ReLU(),
            nn.Linear(18, 36),
            nn.ReLU(),
            nn.Linear(36, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, 28*28),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
learn_rate = 1e-3
epochs = 20
```

```python
# Initilalize model
model = AutoEncoder()  #.to(device)

# MSE loss function
loss_function = nn.MSELoss()

# Adam optimizer
optimizer = torch.optim.Adam(model.parameters(),
                            lr=learn_rate,
                            weight_decay=1e-8)
```

### Train the autoencoder

```python
log_interval = 10
train_losses = []
train_counter = []
test_losses = []
test_counter = [i*len(train_loader.dataset) for i in range(epochs + 1)]
```

```python
def train(epoch):
    model.train() #tell the model it is a training round
    for batch_idx, (x_batch, _) in enumerate(train_loader):
        #print(batch_idx, x_batch.shape, _.shape)
        x_batch = x_batch.reshape(-1, 28*28)  #.to(device)
        pred = model(x_batch)
        loss = loss_function(pred, x_batch)
        loss.backward()
        
        optimizer.step()
        optimizer.zero_grad()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx*len(x_batch), len(train_loader.dataset),
                100.*batch_idx/len(train_loader), loss.item()
            ))
            train_losses.append(loss.item())
            train_counter.append(batch_idx*batch_size_train + epoch*len(train_loader.dataset))

def test():
    model.eval() #tell the model it is being evaluated
    test_loss = 0
    with torch.no_grad():
        for x_batch, _ in test_loader:
            x_batch = x_batch.reshape(-1, 28*28)  #.to(device)
            pred = model(x_batch)
            test_loss += loss_function(pred, x_batch).item()
            val = pred.data.max(1, keepdim=True)[1]
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    print('\nTest set: Avg. loss: {:.4f}'.format(test_loss,))
```

```python
test()
for epoch in range(epochs):
    train(epoch)
    test()
```

```python
plt.plot(train_counter,train_losses,zorder=5)
plt.scatter(test_counter,test_losses,color='r',zorder=15)
plt.style.use('fivethirtyeight')
plt.xlabel('Iterations')
plt.ylabel('Loss')
```

### Compare input and reconstructed data

```python
with torch.no_grad():
    img_batch = next(iter(test_loader))
    img = img_batch[0][0][0]
    reconst = model(img.reshape(-1))
    reconst = reconst.reshape((28,28))
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(1,2,1)
    ax.imshow(img, cmap='grey')
    ax = fig.add_subplot(1,2,2)
    ax.imshow(reconst, cmap='grey')
```
